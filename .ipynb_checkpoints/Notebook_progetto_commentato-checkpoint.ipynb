### 1) Normalize the location field of Users dataset, splitting into city, region, country

```python
users[['City', 'RegionState']] = users['Location'].str.split(',', expand = True, n = 1)
users.drop('Location', axis = 1, inplace = True)
users.head()
```

```python
users[['Region', 'Country']] = users['RegionState'].str.rsplit(',', expand = True, n = 1)
users.drop('RegionState', axis = 1, inplace = True)
users.head()
```

```python
users.loc[(users['Region'] == ' n/a') | (users['Region'] == ' "n/a"'), 'Region'] = np.NaN
```

Abbiamo selezionato la colonna 'Location' del dataset Users. L'obiettivo è quello di dividere questa colonna in tre nuovi attributi, 'City', 'Region' e 'Country'. Per fare ciò, l'idea è l'utilizzo del metodo split. I valori dell'attributo 'Location', tuttavia, non sono normalizzati: la maggior parte delle righe hanno il numero di campi atteso (3) ma tante altre hanno un numero di campi diverso (fino a un massimo di 9 campi). Per risolvere la situazioni, notando che in quasi tutti i casi il primo campo è la città e l'ultimo lo stato, viene usata prima una split con un massimo numero di split pari a 2 per isolare la città e successivamente una rsplit, sempre con massimo numero di split pari a 2, per isolare lo stato. Facendo questo, tutti i campi rimanenti vengono "collassati" nella colonna 'Region'. Le colonne non utili vengono rimosse col metodo drop.

Come ultima cosa si vanno a normalizzare i valori mancanti, rendendoli uguali a np.NaN.

Tuttavia, anche dopo questo procedimento il problema non è risolto completamente. L'attributo 'Region' andrebbe normalizzato, così come gli altri due attributi 'City' e 'Country' andrebbero controllati per verificare la sensatezza dei valori presenti.

### 2) For each book in the Books dataset, compute its average rating

```python
books_w_ratings = pd.merge(books, ratings, on='ISBN', how='left')
books_w_ratings.head()
```

Inizialmente abbiamo combinato i dataset 'Books' e 'Ratings' usando il metodo merge con parametro how=left (per effettuare una outer left join sulla tabella 'Books'). Questa tabella contiene tutti i libri di 'Books' e tutte le valutazioni ad essi corrispondenti (quando presenti). Per il libri senza nessuna valutazione, il valore degli attributi della tabella di destra sono null.

```python
books_avg = books_w_ratings.groupby(['ISBN', 'Book-Title'], as_index=False)['Book-Rating'].mean()
books_avg.head()
```

A questo punto abbiamo effettuato una groupby sugli attributi 'ISBN' e 'Book-Title' e abbiamo generato la media aggregata relativa all'attributo 'Book-Rating', ottenendo per ogni libro la media delle sue valutazioni. Usando l'argomento as_index=False facciamo in modo che gli attributi del raggruppamento non vengano usati come indice del dataset.

### 4) Merge together all rows sharing the same book title, author and publisher. We will call the resulting datset merged books. The books that have not been merged together will not appear in merged books

```python
merged = books.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index=False).size()
merged.head()
```

Abbiamo creato il dataset merged in cui, usando una groupby, abbiamo raggruppato le righe che condividono titolo, autore e publisher. Abbiamo usato il metodo size per ottenere la dimensione di ciascun gruppo, assegnata ad un nuovo attributo omonimo.

```python
merged_books = merged[merged['size'] > 1]
merged_books.head()
```

Utilizzando il fancy indexing abbiamo effettuato un sebsetting del dataset merged mantenendo solo i raggruppamenti di dimensione maggiore o uguale a 2. Il nuovo dataset è stato nominato merged_books come da richiesta.

### 5) For each book in merged books compute its average rating

```python
merged_books_w_avg = pd.merge(merged_books, books_avg, on="Book-Title")
merged_books_w_avg.head()
```

Abbiamo effetuato una combinazione (inner join) del dataset merged_books con il dataset precedentemente ottenuto al punto 2. In questo modo, per ogni raggruppamento, abbiamo ottenuto tutti i rating corrispondenti (alle diverse versioni del libro).

```python
merged_books_w_avg.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index=False)['Book-Rating'].mean()
```

Per ogni raggruppamento di titolo, autore e publisher abbiamo calcolato la media dei rating (su tutte le possibili edizioni dei libri).

### 7) For each book in goodbooks, compute the list of its authors. Assuming that the number of reviews with a text (column work_text_reviews_count) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the shared number of reviews with a text

Ogni elemento della colonna autore è formato dalle stringhe degli autori del libro separati da una virgola. 

Al fine di normalizzare la colonna degli autori:
1. Converto ogni elemento della colonna 'authors' (usando una assign) da stringhe separate da virgola a liste di stringhe, separando le stringhe attraverso una split.

```python
goodbooks_ = goodbooks.assign(authors=goodbooks['authors'].str.split(','))
goodbooks_.head()
```

2. Ottengo una colonna che, per ogni libro, trova il numero delle recensioni testuali divise per il numero di autori del libro corrispondente.

```python
goodbooks_['work_text_reviews_count_per_author'] = goodbooks_['work_text_reviews_count'] / goodbooks_['authors'].str.len()
goodbooks_.head()
```

3. Splitto ogni riga che presenta n autori in n righe separate, dove a ogni libro è associato un solo autore. Questo può essere fatto attraverso il metodo explode che dato un attributo list-like di lunghezza n genera n righe, una per ogni elemento della lista.

```python
goodbooks_normalized = goodbooks_.explode('authors')
goodbooks_normalized.head()
```

4. Trovo il numero totale di recensioni scritte per autore sfruttando la colonna 'work_text_reviews_count_per_author' creata in precedenza. Per facilitare la lettura sono stati passati i nomi delle colonne del nuovo dataset. I valori sono stati arrotondati all'intero più vicino col metodo round. I valori sono stati stampati in ordine alfabetico per facilità di comprensione. 

```python
columns_ = {"authors": "authors", "work_text_reviews_count_per_author": "shared_number_of_reviews_with_a_text"}
authors_reviews_ = pd.DataFrame(goodbooks_normalized.groupby(['authors'], as_index = False)['work_text_reviews_count_per_author']\
                                .sum().round()).rename(columns = columns_)
authors_reviews_.sort_values('authors')
```

Abbiamo valutato anche una soluzione usando un ciclo for che per ogni autore presente nella colonna 'authors' calcolava la somma delle reviews. Veniva generato un dizionario dove le chiavi erano i nomi degli autori e i valori la somma delle reviews.

### 8) For each year of publication, determine the author that has the largest value of the shared number of reviews with a text

L'interpretazione che abbiamo dato alla richiesta è la seguente:

Per ogni anno, abbiamo trovato l'autore (o gli autori), che avendo pubblicato uno o più libri in quell'anno, per cui risulta massima la somma del numero delle recensioni testuali per tutti i libri scritti in quell'anno (considerando che queste vengano equamente divise tra i diversi autori, nel caso questi siano più di uno).

1. Dal dataset goodbooks_normalized abbiamo mantenuto gli autori, l'anno in cui hanno pubblicato e il numero di reviews di ciascun libro (eventualmente condivise con altri autori).

```python
colonne = {'original_publication_year':'original_publication_year', 'authors':'authors', \
           'work_text_reviews_count_per_author':'shared_number_of_reviews_with_a_text_per_year'}
gb_year_author_shared = goodbooks_normalized.groupby(['original_publication_year', 'authors'], as_index=False)\
            ['work_text_reviews_count_per_author'].sum().round().rename(columns = colonne)
gb_year_author_shared.head()
```

2. Abbiamo ottenuto, per ogni anno di pubblicazione, l'indice dell'autore (o degli autori) con numero di recensioni con testo massimo: 

    a) Abbiamo effettuato un raggruppamento sull'attributo 'original_publication_year' di gb_year_author_shared utilizzando il metodo group by
    
    b) Abbiamo applicato il metodo .transform all'attributo 'shared_number_of_reviews_with_a_text_per_year' dell'oggetto grouped, passando un parametro func = max. Il metodo trasform permette di ottenere un dataframe che presenta gli stessi indici del dataframe originale (gb_year_author_shared). L'attributo 'shared_number_of_reviews_with_a_text_per_year' viene però sostituito, per ogni autore di ogni anno, con il valore massimo di quel raggruppamento. Dopodichè viene fatto un confronto, cioè una intersezione fra il dataframe originale e quello trasformato fatta mediante l'operatore ==, ottenendo un vettore booleano nel quale si avrà valore True solo laddove l'autore abbia un numero di review testuali in quell'anno pari al valore massimo dell'anno
    
    c) Usando il fancy indexing, otteniamo solo le righe a cui corrisponde il valore True

```python
idx = gb_year_author_shared.groupby(['original_publication_year'])['shared_number_of_reviews_with_a_text_per_year'].transform(max) == \
                                    gb_year_author_shared['shared_number_of_reviews_with_a_text_per_year']
gb_year_author_shared[idx]
```

### 9) Assuming that there are no errors in the ISBN fields, find the books in both datasets, and compute the difference of average rating according to the ratings and the goodratings datasets

Abbiamo combinato il dataset goodbooks e il dataset books_avg creato al punto 2. Dal dataset goodbooks consideriamo solo gli attributi utili alla merge e l'attributo 'average_rating', mentre dal dataset books_avg consideriamo solo l'attributo 'ISBN' utile alla merge e l'attributo 'Book-Rating'.

```python
both = pd.merge(goodbooks[['book_id', 'isbn', 'authors', 'title', 'average_rating']], books_avg[['ISBN', 'Book-Rating']], 
                left_on = 'isbn', right_on = 'ISBN')
both
```

Come si può vedere di seguito, però, gli attributi 'average_rating' e 'Book-Rating' hanno scale diverse:

```python
print(goodbooks['average_rating'].min(), goodbooks['average_rating'].max())
print(books_avg['Book-Rating'].min(), books_avg['Book-Rating'].max())
```

Pertanto proponiamo di applicare un rescaling dell'attributo 'Book-Rating':

```python
books_avg['Book-Rating_re_scaled'] = books_avg['Book-Rating'] / 2
books_avg.head()
```

E andiamo a riscalare anche il dataset both. Dopodichè andiamo a creare la colonna data dalla differenza dei due valori:

```python
both_re_scaled = pd.merge(goodbooks[['book_id', 'isbn', 'authors', 'title', 'average_rating']], books_avg[['ISBN', 'Book-Rating', 'Book-Rating_re_scaled']], 
                left_on = 'isbn', right_on = 'ISBN')
both_re_scaled['avg_difference'] = both_re_scaled['average_rating'] - both_re_scaled['Book-Rating_re_scaled']
both_re_scaled[['book_id', 'title', 'Book-Rating', 'Book-Rating_re_scaled', 'average_rating', 'avg_difference']].head()
```

La scelta di comprimere la scala di 'Book-Rating' anzichè di espandere quella di 'average_rating' è data dal fatto che espandendo la scala si va a modificare la proporzione fra i voti.

### 10) Split the users dataset according to the age. One dataset contains the users with unknown age, one with age 0-14, one with age 15-24, one with age 25-34, and so on

Come prima cosa andiamo a creare il dataset contenente gli utenti con età sconosciuta:

```python
users_nan = users[users['Age'].isnull()]
users_nan.head()
```

Per separare nelle varie fasce gli utenti, invece, creiamo il dataset degli utenti per cui l'età NON è nulla:

```python
users_temp = users[users['Age'].notnull()]
users_temp.head()
```

Applichiamo il metodo cut sull'attributo 'Age', creando l'attributo 'Age_group' che attribuisce ciascun utente al suo gruppo età. I bin sono stati inseriti manualmente, e gli intervalli sono chiusi a sinistra e aperti a destra:

```python
users_temp['Age_group'] = pd.cut(users_temp['Age'], bins = [0, 15, 25, 35, 45, 55, 65, 75, 85, 95, 1000], right = False,
                                labels = ['<15', '15-24', '25-34', '35-44', '45-54', '55-64', '65-74', '75-84', '85-94', '>95'])
users_temp.head()
```

Per creare i dataset relativi a ciascun gruppo di età seguiamo il seguente procedimento.

Applichiamo tuple() ad un oggetto groupby che crea una tupla dei gruppi. Le tuple, in questo caso, sono composte dal nome del gruppo di età, da una virgola e dall'insieme delle righe appartenenti al gruppo. Dopodichè trasformiamo con dict() la tupla in un dizionario, in modo tale che i nomi dei gruppi diventino le chiavi del dizionario e che il valore associato sia il dataframe composto dagli utenti di quel gruppo:

```python
d = dict(tuple(users_temp.groupby('Age_group')))
d['>95']
```
