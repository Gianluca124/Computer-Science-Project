```python
import pandas as pd
```

### Load datasets

Nella read_csv è stato usato il parametro compression per definire il formato del file compresso (.gzip) contenente i dataset. I primi 3 dataset (Books, Book_Ratings e Users) hanno come separatore il ;, mentre gli ultimi 2 (Goodbooks e Goodbooks_Ratings) sono CSV. Come richiesto è stato passato come encoding l'opzione 'latin-1'.

```python
books = pd.read_csv("https://github.com/gdv/foundationsCS/raw/master/progetti/2021/Books.csv.gz", compression = 'gzip', 
                    sep = ';', encoding = 'latin-1', low_memory=False)
books.head()
```

```python
ratings = pd.read_csv("https://github.com/gdv/foundationsCS/raw/master/progetti/2021/Book-Ratings.csv.gz", compression = 'gzip',
                      sep = ';', encoding = 'latin-1', low_memory=False)
ratings.head()
```

```python
users = pd.read_csv("https://github.com/gdv/foundationsCS/raw/master/progetti/2021/Users.csv.gz", compression = 'gzip', 
                    sep = ';', encoding = 'latin-1', low_memory=False)
users.head()
```

```python
goodbooks = pd.read_csv("https://github.com/gdv/foundationsCS/raw/master/progetti/2021/goodbooks.csv.gz", compression = 'gzip', 
                        encoding = 'latin-1', low_memory=False)
goodbooks.head()
```

```python
goodbooks_ratings = pd.read_csv("https://github.com/gdv/foundationsCS/raw/master/progetti/2021/goodbooks-ratings.csv.gz", 
                                compression = 'gzip', encoding = 'latin-1', low_memory=False)
goodbooks_ratings.head()
```

### 1) Normalize the location field of Users dataset, splitting into city, region, country

```python
users[['City', 'Region', 'Country']] = users['Location'].str.split(',', expand = True, n = 2)
users.drop('Location', axis = 1, inplace = True)
users.head()
```

Abbiamo selezionato la colonna 'Location' del dataset Users. Su di essa abbiamo utilizzato il metodo split separando attraverso la virgola e imponendo un numero massimo di split pari a 2 (quindi generando 3 elementi). Il metodo split genera una lista degli elementi del valore 'Location'; usando il parametro expand gli elementi della lista vengono assegnati alla colonna corrispondente. Abbiamo poi eliminato l'attributo non normalizzato usando il metodo drop.

### 2) For each book in the Books dataset, compute its average rating

```python
books_w_ratings = pd.merge(books, ratings, on='ISBN', how='left')
books_w_ratings.head()
```

Inizialmente abbiamo combinato i dataset 'Books' e 'Ratings' usando il metodo merge con parametro how=left (per effettuare una outer left join sulla tabella 'Books'). Questa tabella contiene tutti i libri di 'Books' e tutte le valutazioni ad essi corrispondenti (quando presenti). Per il libri senza nessuna valutazione, il valore degli attributi della tabella di destra sono null.

```python
books_avg = books_w_ratings.groupby(['ISBN', 'Book-Title'], as_index=False)['Book-Rating'].mean()
books_avg.head()
```

A questo punto abbiamo effettuato una groupby sugli attributi 'ISBN' e 'Book-Title' e abbiamo generato la media aggregata relativa all'attributo 'Book-Rating', ottenendo per ogni libro la media delle sue valutazioni. Usando l'argomento as_index=False facciamo in modo che gli attributi del raggruppamento non vengano usati come indice del dataset.

### 4) Merge together all rows sharing the same book title, author and publisher. We will call the resulting datset merged books. The books that have not been merged together will not appear in merged books

```python
merged = books.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index=False).size()
merged.head()
```

Abbiamo creato il dataset merged in cui, usando una groupby, abbiamo raggruppato le righe che condividono titolo, autore e publisher. Abbiamo usato il metodo size per ottenere la dimensione di ciascun gruppo, assegnata ad un nuovo attributo omonimo.

```python
merged_books = merged[merged['size'] > 1]
merged_books.head()
```

Utilizzando il fancy indexing abbiamo effettuato un sebsetting del dataset merged mantenendo solo i raggruppamenti di dimensione maggiore o uguale a 2. Il nuovo dataset è stato nominato merged_books come da richiesta.

### 5) For each book in merged books compute its average rating

```python
merged_books_w_avg = pd.merge(merged_books, books_avg, on="Book-Title")
merged_books_w_avg.head()
```

Abbiamo effetuato una combinazione (inner join) del dataset merged_books con il dataset precedentemente ottenuto al punto 2. In questo modo, per ogni raggruppamento, abbiamo ottenuto tutti i rating corrispondenti (alle diverse versioni del libro).

```python
merged_books_w_avg.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index=False)['Book-Rating'].mean()
```

Per ogni raggruppamento di titolo, autore e publisher abbiamo calcolato la media dei rating (su tutte le possibili edizioni dei libri).

### 6) For each book in merged books compute the minimum and maximum of the average ratings over all corresponding books in the books dataset

```python
merged_books_w_ratings = pd.merge(merged_books, books_w_ratings, on = ['Book-Title', 'Book-Author', 'Publisher'])
merged_books_w_ratings.head()
```

```python
merged_books_w_ratings.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index=False)['Book-Rating'].max().head()
```

```python
merged_books_w_ratings.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index=False)['Book-Rating'].min().head()
```

Analogamente al punto precedente, abbiamo combinato il dataset merged_books con il dataset books_w_ratings ottenendo un dataset in cui, per ciascun raggruppamento di titolo, autore e publisher vengono riportati tutti i ratings. Usando i metodi min e max, dopo aver effettuato una groupby, vengono riportati il rating massimo e minimo per ciascun raggruppamento.

### 7) For each book in goodbooks, compute the list of its authors. Assuming that the number of reviews with a text (column work_text_reviews_count) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the shared number of reviews with a text

Ogni elemento della colonna autore è formato dalle stringhe degli autori del libro separati da una virgola. 

Al fine di normalizzare la colonna degli autori:
1. Converto ogni elemento della colonna 'authors' (usando una assign) da stringhe separate da virgola a liste di stringhe, separando le stringhe attraverso una split.

```python
goodbooks_ = goodbooks.assign(authors=goodbooks['authors'].str.split(','))
goodbooks_.head()
```

2. Ottengo una colonna che, per ogni libro, trova il numero delle recensioni testuali divise per il numero di autori del libro corrispondente.

```python
goodbooks_['work_text_reviews_count_per_author'] = goodbooks_['work_text_reviews_count'] / goodbooks_['authors'].str.len()
goodbooks_.head()
```

3. Splitto ogni riga che presenta n autori in n righe separate, dove a ogni libro è associato un solo autore. Questo può essere fatto attraverso il metodo explode che dato un attributo list-like di lunghezza n genera n righe, una per ogni elemento della lista.

```python
goodbooks_normalized = goodbooks_.explode('authors')
goodbooks_normalized.head()
```

4. Trovo il numero totale di recensioni scritte per autore sfruttando la colonna 'work_text_reviews_count_per_author' creata in precedenza. Per facilitare la lettura sono stati passati i nomi delle colonne del nuovo dataset. I valori sono stati arrotondati all'intero più vicino col metodo round. I valori sono stati stampati in ordine alfabetico per facilità di comprensione. 

```python
columns_ = {"authors": "authors", "work_text_reviews_count_per_author": "shared_number_of_reviews_with_a_text"}
authors_reviews_ = pd.DataFrame(goodbooks_normalized.groupby(['authors'], as_index = False)['work_text_reviews_count_per_author']\
                                .sum().round()).rename(columns = columns_)
authors_reviews_.sort_values('authors')
```

Abbiamo valutato anche una soluzione usando un ciclo for che per ogni autore presente nella colonna 'authors' calcolava la somma delle reviews. Veniva generato un dizionario dove le chiavi erano i nomi degli autori e i valori la somma delle reviews.

### 8) For each year of publication, determine the author that has the largest value of the shared number of reviews with a text

L'interpretazione che abbiamo dato alla richiesta è la seguente:

Per ogni anno, trovare l'autore che, sommando il numero delle recensioni per tutti i libri scritti in quell'anno ha il numero massimo.

Dal dataset goodbooks_normalized abbiamo mantenuto gli autori, l'anno in cui hanno pubblicato e il numero di reviews di ciascun libro (eventualmente condivise con altri autori).

```python
colonne = {'original_publication_year':'original_publication_year', 'authors':'authors', \
           'work_text_reviews_count_per_author':'shared_number_of_reviews_with_a_text'}
gb_year_author_shared = goodbooks_normalized.groupby(['original_publication_year', 'authors'], as_index=False)\
            ['work_text_reviews_count_per_author'].sum().round().rename(columns = colonne)
gb_year_author_shared.head()
```

Ora seleziono, per ogni anno, gli autori con shared massimo utilizzando una groupby per anno ed autore e una iloc per ottenere le righe il cui indice è stato individuato dal metodo idxmax.

```python
best_author_of_the_year = gb_year_author_shared.iloc[gb_year_author_shared.groupby(['original_publication_year'])\
                                   ['shared_number_of_reviews_with_a_text'].idxmax()]
best_author_of_the_year.head()
```

```python
best_author_of_the_year = gb_year_author_shared.iloc[gb_year_author_shared.groupby(['original_publication_year'])\
                                   ['shared_number_of_reviews_with_a_text'].idxmax()]
best_author_of_the_year.head()
```

```python
idx = gb_year_author_shared.groupby(['original_publication_year'])['shared_number_of_reviews_with_a_text'].transform(max) == \
                                    gb_year_author_shared['shared_number_of_reviews_with_a_text']
gb_year_author_shared[idx]
```
